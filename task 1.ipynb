```python
import pandas as pd

# Load the Titanic dataset from a URL instead of a local file
# This uses the Titanic dataset from Seaborn's data repository
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# See the first 5 rows
print(data.head())

# Check missing values
print(data.isnull().sum())
```

```python
# Instead of using inplace=True with chained assignments, use the recommended approaches
# Method 1: Assign the result back to the column
data['Age'] = data['Age'].fillna(data['Age'].median())

# Method 2: Use the DataFrame.fillna method directly
data = data.fillna({'Embarked': data['Embarked'].mode()[0]})

# Drop 'Cabin' column only if it exists
if 'Cabin' in data.columns:
    data.drop('Cabin', axis=1, inplace=True)

# Verify changes
print("Missing values after cleaning:")
print(data.isnull().sum())
```

```python
# Convert 'Sex' to numbers
data['Sex'] = data['Sex'].map({'male': 1, 'female': 0})

# Convert 'Embarked' into 0/1 columns
data = pd.get_dummies(data, columns=['Embarked'], drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['Age', 'Fare']] = scaler.fit_transform(data[['Age', 'Fare']])
```

```python
print(data.head())
print(data.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for 'Fare'
sns.boxplot(data['Fare'])
plt.show()
```
